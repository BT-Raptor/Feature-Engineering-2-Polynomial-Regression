Load the libaries & Data

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
np.set_printoptions(suppress=True)

all_data=pd.read_csv('opp.csv')
all_data=all_data.drop(['case','site','Pop','sex'],axis=1)
all_data.head()
```

```python
all_data=all_data.dropna()
temp_df=all_data.copy(deep=True)
temp_df['hdlngth_2']=temp_df['hdlngth']**2
temp_df['skullw_2']=temp_df['skullw']**2
temp_df['totlngth_2']=temp_df['totlngth']**2
temp_df['taill_2']=temp_df['taill']**2
temp_df['footlgth_2']=temp_df['footlgth']**2
temp_df['earconch_2']=temp_df['earconch']**2
temp_df['eye_2']=temp_df['eye']**2
temp_df['chest_2']=temp_df['chest']**2
temp_df['belly_2']=temp_df['belly']**2
```

```python
Y=all_data['age']
X=all_data.drop(['age'],axis=1)
model=LinearRegression().fit(X,Y)
all_data['preds']=np.round(model.predict(X))
```

```python
Y=temp_df['age']
X=temp_df.drop(['age'],axis=1)
model=LinearRegression().fit(X,Y)
all_data['new_preds']=np.round(model.predict(X))
all_data=all_data[['age','preds','new_preds']]
all_data.head(15)
```
